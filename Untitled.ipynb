```python
import matplotlib.pyplot as plt
import torch
import torch.utils.data as data
import zuko
from scipy.stats import norm
import numpy as np
```

```python
def two_moons(n: int, sigma: float = 1e-1):
    theta = 2 * torch.pi * torch.rand(n)
    label = (theta > torch.pi).float()

    x = torch.stack((
        torch.cos(theta) + label - 1 / 2,
        torch.sin(theta) + label / 2 - 1 / 4,
    ), axis=-1)

    return torch.normal(x, sigma), label

raw_ds = two_moons(16384)
```

```python
split = 1024
```

```python
unlabeled = raw_ds[0][:-split], raw_ds[1][:-split]
labeled = raw_ds[0][-split:], raw_ds[1][-split:]
```

```python
whole_ds = data.TensorDataset(*raw_ds)
labeled_set =  data.TensorDataset(*labeled)
unlabeled_set =  data.TensorDataset(*unlabeled)

wholeloader = data.DataLoader(whole_ds, batch_size=64, shuffle=True)
labeledloader = data.DataLoader(labeled_set, batch_size=64, shuffle=True)
unlabeled = data.DataLoader(unlabeled_set, batch_size=64, shuffle=True)
```

```python
whole_nf = zuko.flows.NSF(features=2, transforms=3, hidden_features=(64, 64))
whole_optimizer = torch.optim.Adam(whole_nf.parameters(), lr=1e-3)

for epoch in range(20):
    losses = []

    for x, label in wholeloader:
        loss = -whole_nf().log_prob(x).mean()
        loss.backward()

        whole_optimizer.step()
        for param in whole_nf.parameters():
            param.grad = None


        losses.append(loss.detach())

    losses = torch.stack(losses)

    print(f'({epoch})', losses.mean().item(), '±', losses.std().item())
```

```python
labeled_nf = zuko.flows.NSF(features=2, transforms=3, hidden_features=(64, 64))
labeled_optimizer = torch.optim.Adam(labeled_nf.parameters(), lr=1e-3)
for epoch in range(1000):
    losses = []

    for x, label in labeledloader:
        loss = -labeled_nf().log_prob(x).mean()
        loss.backward()

        labeled_optimizer.step()
        for param in labeled_nf.parameters():
            param.grad = None


        losses.append(loss.detach())

    losses = torch.stack(losses)

    print(f'({epoch})', losses.mean().item(), '±', losses.std().item())
```

```python
extra_factor = lambda m : (2*np.pi)**(-(m-1)/2)

def get_threshold(prob,m):
    x = norm.ppf((prob+1)/2) #idcf
    prob_norm = norm.pdf(x) #pdf
    em = extra_factor(m) #scale for more dimensions
    return em*prob_norm

get_threshold(0.95, 2)

threshhold_95 = np.exp(-norm.ppf(0.95)**2/2)/(2*torch.pi)
threshhold_95
```

```python
threshhold_95=0.07375
```

```python
with torch.no_grad():
    whole_nf_probs = torch.exp(whole_nf().log_prob(unlabeled_set.tensors[0]))
    labeled_nf_probs = torch.exp(labeled_nf().log_prob(unlabeled_set.tensors[0]))
```

```python
(whole_nf_probs > threshhold_95).sum()/whole_nf_probs.shape[0]
```

```python
(labeled_nf_probs > threshhold_95).sum()/labeled_nf_probs.shape[0]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10), sharex=True, sharey=True)
s=2
ax[0].scatter(*labeled_set.tensors[0].T, label="Labeled Point", s=s)
ax[0].scatter(*unlabeled_set.tensors[0][(labeled_nf_probs < threshhold_95)].T,  label="P_ds(U) < Given threshold", s=s)
ax[1].scatter(*whole_ds.tensors[0].T, label="Instances which NF was trained on", s=s)
ax[1].scatter(*unlabeled_set.tensors[0][(whole_nf_probs < threshhold_95)].T,  label="P_ds(U) < Given threshold", s=s)
plt.legend()
ax[0].set_title("$\mathcal{L} \in \mathbb{R}^{1024 x m}$")
ax[1].set_title("$\mathcal{D} \in \mathbb{R}^{16384 x m}$")

plt.show()
```
